# Gráficos em Python - Matplotlib

A principal biblioteca para gerar gráficos científicos em Python é a Matplotlib (do projeto SciPy).

Essa biblioteca tem diversas formas de uso. A mais geral é através de seu módulo `matplotlib.pyplot`, que convencionalmente é importado com o nome `plt`.

Quando estamos trabalhando num notebook Jupyter (ou no qtconsole), é comum desejarmos que os gráficos apareçam *inline*, isto é, logo a seguir dos comandos que os geram (como as saídas de comandos de texto).

Portanto, geralemente ao usar Matplotlib em um notebook, começamos com:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

A Matplotlib trabalha junto com o NumPy. Portanto, vamos também importar `numpy`.

```python
import numpy as np
```

## Gráfico de barras simples

No nosso primeiro exemplo, vemos um gráfico de barras simples. Esse tipo de gráfico é muito usado para visualizações quantitativas

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.bar(languages,students)
plt.show()
```

## Gráfico de barras agrupadas

No nosso segundo exemplo, vemos um gráfico de barras agrupadas, usado para visualizações quantitativas em conjunto.

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
```

## Gráfico de barras empilhadas

No nosso terceiro exemplo, vemos um gráfico de barras empilhadas, usado para visualizações quantitativas em conjunto.

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, men_means, width, yerr=men_std, label='Men')
ax.bar(labels, women_means, width, yerr=women_std, bottom=men_means,
       label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```

## Gráfico de linhas e curvas

No nosso quarto exemplo, vemos um gráfico de linha e curvas, também usado para visualizações quantitativas em conjunto.

```python
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

## Gráficos - Comparações

Temos aqui uma comparação com tipos diferentes de gráficos para uma mesma quantidade de valores. Qual a melhora para se usar? Depende do que você quer!

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(12, 5))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```
